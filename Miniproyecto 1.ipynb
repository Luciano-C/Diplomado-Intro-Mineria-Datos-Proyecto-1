```python
# Primero se limpiaran datos de columna SEX. Columna SEX_y tiene números pero está definida como "object". Se procede a convertirla.
df1_2_3_4["SEX_y"] = pd.to_numeric(df1_2_3_4["SEX_y"], errors="coerce")
```

```python
# Hay 6 valores que no se tienen en la columna SEX_y. Se intentará buscarlos en la columna SEX_x y reemplazarlos
df1_2_3_4["SEX_y"].isna().sum()
```

```python
# Se utilizó los valores de la columna "SEX_y" para reemplazar los faltantes en la columna "SEX_x" según el criterio:
# 1 corresponde a "male", 2 corresponde a "female". Esto se dedujo al comparar filas presentes en ambas columnas.

for i in range (0, len(df1_2_3_4)):
    if pd.isna(df1_2_3_4.loc[i,"SEX_x"]):
        if df1_2_3_4.loc[i, "SEX_y"] == 1:
            df1_2_3_4.loc[i, "SEX_x"] = "male"
        elif df1_2_3_4.loc[i, "SEX_y"] == 2:
            df1_2_3_4.loc[i, "SEX_x"] = "female"

# No quedan valores vacíos en SEX_x            
df1_2_3_4["SEX_x"].isna().sum()

```

```python
# Para verificar que ambas columnas sean consistentes:
# - Se reemplazan los 1 por "male" y los 2 por "female" en "SEX_y"
# - Se reemplazan los valores faltantes en "SEX_y" por los correspondientes en "SEX_x"
# Luego se comparan y al ser iguales se puede eliminar una, en este caso se eliminará "SEX_y" y se renombrará "SEX_x" más adelante.

df1_2_3_4['SEX_y'] = df1_2_3_4['SEX_y'].replace(1,"male")
df1_2_3_4['SEX_y'] = df1_2_3_4['SEX_y'].replace(2,"female")
for i in range (0, len(df1_2_3_4)):
    if pd.isna(df1_2_3_4.loc[i,"SEX_y"]):
        df1_2_3_4.loc[i,"SEX_y"] = df1_2_3_4.loc[i,"SEX_x"]

df1_2_3_4["SEX_x"].equals(df1_2_3_4["SEX_y"])
```

```python
if "SEX_y" in df1_2_3_4.columns:
    df1_2_3_4.drop("SEX_y", inplace=True, axis=1)

```

<h1>Normalización</h1>

```python
# Se crea un dataframe de respaldo para no perder los datos originales.
# Se decidió aplicar una normalización z score a las columnas "Limit Balance", todas las columnas "BILL AMT" y todas las columnas
# "PAY AMT". Esto porque tienen valores dispares que varían bastante, pueden ser negativos, 0 o cientos de miles.
# Se descartó normalización decimal o lineal porque al existir límites tan dispares con respecto a algunos valores, se obtendrían
# algunos valores con muchos decimales al aplicar la normalización.
df_final_normalizado = df_final.copy()
```

```python
columnas_a_normalizar = ['Limit Balance', 'BILL AMT 1', 'BILL AMT 2', 'BILL AMT 3', 'BILL AMT 4', 'BILL AMT 5',
                         'BILL AMT 6', 'PAY AMT 1','PAY AMT 2', 'PAY AMT 3', 'PAY AMT 4', 'PAY AMT 5', 'PAY AMT 6']

# Se aplica la fórmula de z score para las columnas seleccionadas 1 grado de libertad (ddof)
df_final_normalizado[columnas_a_normalizar] = (df_final_normalizado[columnas_a_normalizar]-df_final_normalizado[columnas_a_normalizar].mean())/df_final_normalizado[columnas_a_normalizar].std(ddof=1)
df_final_normalizado.head()

```
